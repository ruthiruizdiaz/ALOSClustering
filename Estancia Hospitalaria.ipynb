```python
# Filtrar el año 2016 y encontrar los últimos cinco años
df_filtrado = df[df['time'] != 2016]
ultimos_cinco_anos = sorted(df_filtrado['time'].unique(), reverse=True)[:5]
df_filtrado = df_filtrado[df_filtrado['time'].isin(ultimos_cinco_anos)]

# Agrupar por país y año, y calcular la suma de los valores
df_agrupado = df_filtrado.groupby(['time', 'location'])['value'].sum().unstack()

# Mostrar los resultados
print(df_agrupado)

# Generar el gráfico de barras
df_agrupado.plot(kind='bar', figsize=(12, 8), colormap='tab20')

plt.title('Valores por país para los últimos 5 años (sin incluir 2016)')
plt.xlabel('Año')
plt.ylabel('Valor')
plt.legend(title='País')
plt.grid(True)
plt.show()
```

```python
# Calcular la correlación entre el año y el valor
correlacion = df['time'].corr(df['value'])
print(f'Coeficiente de correlación de Pearson: {correlacion}')

# Generar el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df['time'], df['value'], c='blue', alpha=0.5)
plt.title('Dispersión entre Año y Valor')
plt.xlabel('Año')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df.describe()
```

```python
df1= df[['time','value']]
df1
```

```python
#Calcular la cantidad optima de clusters con el método codo de Jambu
wcss= []
for i in range (1,11):
  kmeans= KMeans(n_clusters= i, max_iter=300)
  kmeans.fit(df1) #aplicando kmeans a la base de datos
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("Codo de Jambu")
plt.xlabel('Numero de clusters')
plt.ylabel('WCSS') #Es un indicador de que tan similares son los individuos dentro de los clusters
plt.show()
```

```python
#Clustering 1, con datos originales
clustering= KMeans(n_clusters=3,max_iter=300) #crea el modelo
clustering.fit(df1) #aplica el modelo a la base de datos
```

```python
df1['KMeans_Clusters']=clustering.labels_ #los resultados del clustering se guardan en labels_ del modelo
df1
```

```python
fig= plt.figure(figsize=(6,6))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('Componente 1',fontsize=15)
ax.set_ylabel('Componente 2',fontsize=15)
ax.set_title('Componentes principales',fontsize=20)
color_theme=py.array(['blue','green','orange'])
ax.scatter(x=df1['time'],y=df1['value'],c=color_theme[df1['KMeans_Clusters']],s=50)
plt.show()
```

```python
#Clustering 2 con los datos ajustados
# Usar pivot_table en lugar de pivot para manejar duplicados
df_pivot = df.pivot_table(index='location', columns='time', values='value', aggfunc='mean')

# Llenar los valores faltantes con NaN si es necesario
df_pivot = df_pivot.reset_index()

# Normalizar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_pivot.drop(columns=['location']))

# Añadir la columna de location de nuevo
df_pivot_scaled = pd.DataFrame(df_scaled, columns=df_pivot.columns[1:], index=df_pivot['location'])

# Mostrar el DataFrame pivoteado
print(df_pivot)
```

```python
# Guardar en df2 todos los datos de df_pivot pero sin la columna location
df2 = df_pivot.drop(columns=['location'])
# Eliminar columnas (años) que tengan algún NaN
df2 = df2.dropna(axis=1)
# Mostrar el DataFrame df2
print(df2)
```

```python
clustering= KMeans(n_clusters=3,max_iter=300) #crea el modelo
clustering.fit(df2) #aplica el modelo a la base de datos
```

```python

df2['KMeans_Clusters']=clustering.labels_ #los resultados del clustering se guardan en labels_ del modelo
df2
```

```python
from sklearn.decomposition import PCA

# Convert all column names to strings
df2.columns = df2.columns.astype(str)

# Now apply PCA
pca = PCA(n_components=2)  # Initialize PCA with 2 components
pca_result = pca.fit_transform(df2)  # Apply PCA

# Create new columns in df2 for PCA results
df2['PCA1'] = pca_result[:, 0]
df2['PCA2'] = pca_result[:, 1]
```

```python
fig= plt.figure(figsize=(6,6))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('Componente 1',fontsize=15)
ax.set_ylabel('Componente 2',fontsize=15)
ax.set_title('Componentes principales',fontsize=20)
color_theme=py.array(['blue','green','orange'])
ax.scatter(x=df2['PCA1'],y=df2['PCA2'],c=color_theme[df2['KMeans_Clusters']],s=50)
#ax.scatter(x=hola['time'],y=hola['value'],c=color_theme[hola['KMeans_Clusters']],s=50)
plt.show()
```
